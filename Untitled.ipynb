```python
from PIL import Image
from IPython.display import Image as inlineimage
import io
import os
import numpy as np
import imageio
from natsort import natsorted
```

```python
availfiles = os.listdir('movieslices')
slicefiles = [x for x in availfiles if x.endswith('.jpg')]
images = []
endframe=70
maxsize = (200, 200)
ii = 0
for filename in natsorted(slicefiles):
    if ii > endframe:
        break
    img = Image.open('movieslices/' + filename)
    w, h = img.size
    cropped_img = img.crop((0,35,650,h-35))
    maxsize = (200, 200)
    cropped_img.thumbnail(maxsize, Image.ANTIALIAS)
    images.append(np.array(cropped_img))
    ii += 1
imageio.mimsave('movie.gif', images)
```
